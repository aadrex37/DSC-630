```python
df=pd.read_excel("Wages and salaries for Private industry workers in All industries and occupations, 12-month percent change.xlsx")
df
##import df 
```

```python
df['Estimate Value'].describe()
##describes estimated value and other descriptive statistics
```

```python
df['Standard Error'].describe()
##describes standard error and other descriptive statistics
```

```python
plt.hist(df['Estimate Value'], bins=40)
##Creates histogram
plt.xlabel('Estimated Value')
##defines x-axis
plt.ylabel('Count')
##defines y-axis
plt.title("Wages and salaries for Private industry workers in All industries and occupations")
##Title of graph
plt.show()
##displays bar graph
```

```python
df.boxplot(column=['Estimate Value'])##create boxplot
```

```python
df.to_csv("Wages and salaries for Private industry workers.csv")##export as csv
```

```python
### Bivariate Plots
```

```python
plt.scatter(df['Year'],df['Estimate Value'])
##plots scatter plot for valuer for each quarter by each year
plt.xlabel('Year')
##defines x-axis
plt.ylabel('Estimated Value')
##defines y-axis
plt.title("quarterly Wages and salaries for Private industry workers by year")
##Title of graph
plt.show()
##displays graph
```

```python
plt.matshow(df.corr())
plt.show()
##correlation matrix
```

```python
table = pd.pivot_table(df,index=['Year','Period'])
table
##creates pivot table
```
