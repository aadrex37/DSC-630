```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dodgers.csv")
df
```

```python
pd.set_option("display.max_rows", None)
df
```

```python
df.boxplot(column=['attend'])
plt.show()
## we want to focus on the lowest 25% of attendance days to see which days need a theme night
##This shows that we want to focus on days with attenance below 34000
```

```python
##New df with attendance below 35,000 (low 25%) and no other night specials
df1 = df[(df['attend']<=35000) & (df['cap']=='NO')& (df['shirt']=='NO')& (df['fireworks']=='NO')& (df['bobblehead']=='NO')]
df1
```

```python
df1=df1.reset_index()
df1
##19 potential days
```

```python
x=df1['month']
y=df1['day_of_week']
plt.scatter(x=x,y=y,c='green')##plots scatter plot mix of months and days
plt.xlabel('Month')
plt.ylabel('Day of the week')
plt.title('Distribution of low attendance days by day of the week and month of the year')
plt.show()
##days to focus on are May, Mondays, and September(important games at end of year)
```

```python
df2= df1[(df1['day_of_week']=='Monday') | (df1['month']=='MAY') | (df1['month']=='SEP')]
df2
```

```python
x=df2['temp']
y=df2['attend']
plt.scatter(x=x,y=y,c='green')##plots scatter plot mix of months and days
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
plt.xlabel('Tempurature')
plt.ylabel('Attendance')
plt.title('Distribution of low attendance days by tempurature with regression line')
plt.show()
```
